```python
#Importing libraries that we need

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Reading the data

df = pd.read_csv("Titanic-Dataset.csv")


sns.set(style="whitegrid", palette="muted")

#distribution of sexulity by if they survived

plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Survival Distribution by Gender")
plt.show()

#distribution of survived by their boat class

plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Pclass", hue="Survived")
plt.title("Survival Distribution by Passenger Class")
plt.show()

#box plot of those who survived and those who did not survive


plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="Survived", y="Age")
plt.title("Age Distribution by Survival")
plt.show()

#distribution of survivers by their fare share

plt.figure(figsize=(6,5))
sns.kdeplot(data=df, x="Fare", hue="Survived", fill=True)
plt.title("Fare Distribution by Survival")
plt.show()

#distribution by how many siblings where with them(sibsp)


plt.figure(figsize=(6,5))
sns.countplot(data=df, x="SibSp", hue="Survived")
plt.title("Survival vs Number of Siblings/Spouses Aboard")
plt.show()

#distribution of how many parents or children they had (parch) based on if they survived or not

plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Parch", hue="Survived")
plt.title("Survival vs Number of Parents/Children Aboard")
plt.show()

#some instances for pairploting them

sns.pairplot(df[['Survived','Pclass','Age','Fare']], hue="Survived", diag_kind="kde")
plt.suptitle("Pairplot of Survival with Age, Fare, and Class", y=1.02)
plt.show()

#vilion plotting69


plt.figure(figsize=(8,6))
sns.violinplot(data=df, x="Pclass", y="Age", hue="Survived", split=True)
plt.title("Age Distribution by Class and Survival")
plt.show()

#researching who survived base on the transportation

plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Embarked", hue="Survived")
plt.title("Survival Distribution by Embarkation Port")
plt.show()

#creating hitmap

plt.figure(figsize=(8,6))
sns.heatmap(df[['Survived','Pclass','Age','Fare','SibSp','Parch']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

#distribution based on age and fare of if they survived

plt.figure(figsize=(7,6))
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived", style="Sex")
plt.title("Age vs Fare by Survival and Gender")
plt.show()

#distribution of how they survived related to sex and class

pd.crosstab(df['Pclass'], df['Sex']).plot(kind='bar', stacked=True, figsize=(7,6))
plt.title("Class Distribution by Gender")
plt.show()

#how many people stay alive base on they class

plt.figure(figsize=(7,6))
sns.barplot(data=df, x="Pclass", y="Survived", hue="Sex")
plt.title("Survival Rate by Class and Gender")
plt.show()

#box plot for fare and class

plt.figure(figsize=(7,6))
sns.boxplot(data=df, x="Pclass", y="Fare", hue="Survived")
plt.title("Fare Distribution by Class and Survival")
plt.show()

#histogram of age and how many stayed alive

plt.figure(figsize=(7,6))
sns.histplot(data=df, x="Age", hue="Survived", bins=20, kde=True)
plt.title("Age Histogram by Survival")
plt.show()
```
